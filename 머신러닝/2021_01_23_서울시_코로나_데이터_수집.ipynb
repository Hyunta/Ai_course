```
all_confirm_patient = [ ]

for page in range(1,101):
  print("page =",page)
  start_index = (page-1)*100
  print("start_index =",start_index)

  url = f"https://news.seoul.go.kr/api/27/getCorona19Status/get_status_ajax_pre.php?draw={page}&start={start_index}&length=100&_=1611628635795"
  print("url =",url)
  response = requests.get(url)
  print("response = ",response)

  print("*"*30)
  print("response.json()")
  print("*"*30)
  print(response.json())
  print("*"*30)
  covid19 = response.json()
  print("covid19")
  print("*"*30)
  print(covid19)
  print("*"*30)
  print("covid19['data']")
  print("*"*30)
  print(covid19["data"])
  confirm_patient = covid19["data"]
  print("confirm_patient")
  print("*"*30)
  print(confirm_patient)
  print("*"*30)
  print("확진자 정보를 추가 했어요")
  all_confirm_patient.append(pd.DataFrame(confirm_patient))
  print("all_confitm_patient=", all_confirm_patient)
  print("*"*30)
  time.sleep(0.5)
```
