```
from sklearn.datasets import load_breast_cancer
```

```
load_breast_cancer()
```

```
dataset = load_breast_cancer()
```

```
dataset
```

```
dataset.data
```

```
X_features = dataset.data
```

```
y_label = dataset.target
```

```
feature_names = dataset.feature_names
```

```
import pandas as pd
```

```
cancer_df = pd.DataFrame(data = X_features, columns= feature_names)
```

```
cancer_df["target"] = y_label
```

```
from sklearn.model_selection import train_test_split
train_test_split(X_features, y_label)
```

```
X_train, X_test, y_train, y_test = train_test_split(X_features, y_label)
```

```
from xgboost import XGBClassifier
```

```
xgboost = XGBClassifier()
```

```
xgboost.fit(
            X_train,
            y_train
)
```

```
prediction = xgboost.predict(X_test)
```

```
prediction == y_test
```

```
(y_test == prediction).sum()
```

```
(y_test == prediction).shape[0]
```

```
(y_test == prediction).sum() / (y_test == prediction).shape[0]
```

```
from sklearn.metrics import accuracy_score
```

```
accuracy_score(y_test, prediction)
```

```
params = {
    "max_depth" : [5,7,9],
    "min_child_weight" : [3,5,7,9],
    "colsample_bytree" : [0.5, 0.75, 1.0],
    "learning_rate" : [0.01, 0.1, 0.2, 0.3, 0.5]
}
```

```
from sklearn.model_selection import GridSearchCV
```

```
xgb_clf = XGBClassifier()
```

```
gridcv = GridSearchCV(xgb_clf, param_grid= params)
```

```
gridcv.fit(
    X_train,
    y_train,
    eval_set = [(X_test, y_test)]
)
```

```
gridcv.best_params_
```

```
best_xgboost = gridcv.best_estimator_
```

```
prediction = best_xgboost.predict(X_test)
```

```
y_test == prediction
```

```
(y_test == prediction).sum()
```

```
(y_test == prediction).shape[0]
```

```
(y_test == prediction).sum()/(y_test == prediction).shape[0]
```

```
best_xgboost.feature_importances_
```

```
importance = pd.Series(index=feature_names, data=best_xgboost.feature_importances_)
```

```
importance.sort_values(ascending=False)[0:10]
```
