```
pd.DataFrame(
    {
     "년월주":week_confirm.index,
     "확진자수":week_confirm.values

    }
)
```

```
df_week_confirm = pd.DataFrame(
    {
     "년월주":week_confirm.index,
     "확진자수":week_confirm.values

    }
)
```

```
df_week_confirm.sort_values(["년월주"], inplace=True)
```

```
df_week_confirm["누적확진자수"] = df_week_confirm["확진자수"].cumsum()
```

```
df_week_confirm.head(2)
```

```
fig, ax = plt.subplots (figsize = (20, 10))
sns.barplot(data=df_week_confirm, x="년월주", y="누적확진자수")
fig.autofmt_xdate(rotation=90)
```

```
old_year = 0
old_month = 0

for i in range(df_week_confirm.shape[0]):
  리스트데이터 = ""
  print("i=",i)
  print("df_week_confirm.iloc[",i,"][0]",df_week_confirm.iloc[i][0])

  print("(df_week_confirm.iloc[",i,"][0])[0:2]= ",(df_week_confirm.iloc[i][0])[0:2] )
  year = int((df_week_confirm.iloc[i][0])[0:2])
  print("year=",year)
  print("old_year=",old_year)
  if year != old_year:
    print("*"*30)
    print("년도 변화!!!")
    print("old_year 를 year 로 수정")
    old_year = year
    print("수정된 old_year=",old_year)
    리스트데이터 += (df_week_confirm.iloc[i][0])[0:2]+"/"
    print("리스트데이터에 연도 추가 리스트데이터=",리스트데이터)
    print("*"*30)
  
  print("(df_week_confirm.iloc[",i,"][0])[3:5]= ",(df_week_confirm.iloc[i][0])[3:5] )
  month = int((df_week_confirm.iloc[i][0])[3:5])
  print("month=",month)
  if month != old_month:
    print("*"*30)
    print("월 변화!!!")
    print("old_month 를 month 로 수정")
    old_month = month
    print("수정된 old_month=",old_month)
    리스트데이터 += (df_week_confirm.iloc[i][0])[3:5]
    print("리스트데이터에 월 추가 리스트데이터=",리스트데이터)
    print("*"*30)

  print("(df_week_confirm.iloc[",i,"][0])[6:8]= ",(df_week_confirm.iloc[i][0])[6:8] )
  리스트데이터 += "(" +(df_week_confirm.iloc[i][0])[6:8]+")"
  print("리스트데이터에 주 추가 리스트데이터=",리스트데이터)
  week = int((df_week_confirm.iloc[i][0])[6:8])
  print("week=",week)
  print("최종 리스트데이터=", 리스트데이터)
  print("="*30)
```

```
old_year = 0
old_month = 0

x_labels=[]
for i in range(df_week_confirm.shape[0]):
  리스트데이터 = ""

  year = int((df_week_confirm.iloc[i][0])[0:2])
  if year != old_year:
    old_year = year
    리스트데이터 += (df_week_confirm.iloc[i][0])[0:2]+"/"

  month = int((df_week_confirm.iloc[i][0])[3:5])
  if month != old_month:
    old_month = month
    리스트데이터 += (df_week_confirm.iloc[i][0])[3:5]

  리스트데이터 += "(" +(df_week_confirm.iloc[i][0])[6:8]+")"

  x_labels.append(리스트데이터)

```

```
x_labels
```

```
fig, ax = plt.subplots (figsize=(20,10))
sns.barplot(data=df_week_confirm, x="년월주", y="누적확진자수")
fig.autofmt_xdate()
ax.set_xticklabels(labels = x_labels , fontdict={"size": 9})
```
